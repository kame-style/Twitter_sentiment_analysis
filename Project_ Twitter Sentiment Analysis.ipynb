```python
docs=[]

for i in range(df.shape[0]) :
  doc=df.iloc[i,3]
  docs.append(doc)

docs_test=[]

for i in range(df_test.shape[0]) :
  doc=df_test.iloc[i,2]
  docs_test.append(doc)

print(docs[0])
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
nltk.download('punkt')                        ## need to download bec it gives error punkt not found etc..
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from nltk.corpus import wordnet
nltk.download('wordnet')
def get_simple_tag(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else :
    return wordnet.NOUN

```

```python
##cleaning the docs
stop_words = stopwords.words('english')


import string
punctuations=string.punctuation
stop_words= stop_words + list(punctuations)

word_docs=[]

for doc in docs :
  words = word_tokenize(doc)
  clean_words = [w.lower() for w in words if not w.lower() in stop_words]
  word_docs.append(clean_words)

word_docs_test=[]
for doc in docs_test :
  words = word_tokenize(doc)
  clean_words = [w.lower() for w in words if not w.lower() in stop_words]
  word_docs_test.append(clean_words)

lemmatizer=WordNetLemmatizer()


def clean_docs(doc):
  output=[]
  for w in doc:
    pos = pos_tag([w]) ## IMP pass word as array . otherwise it will give tag for individual letters
    clean_word=lemmatizer.lemmatize(w,pos=get_simple_tag(pos[0][1])) ## pos is like [('word','pos tag')] since we need to pass pos tag hence we need to give [0][1]
    output.append(clean_word.lower())
  return output

final_docs=[]

for doc in word_docs:
     final_docs.append(clean_docs(doc))

final_docs_test=[]

for doc in word_docs_test:
     final_docs_test.append(clean_docs(doc))

```

```python
# building feature set

train_docs=final_docs
test_docs=final_docs_test
all_words=[]

for doc in train_docs:
  for w in doc:
    all_words.append(w)

freq=nltk.FreqDist(all_words)
common_words=freq.most_common(5000)

features=[i[0] for i in common_words]


def get_feature_dict(doc):      ## to mark if the word in dictionary is present in the doc by true or false
  curr_features={}
  words_set=set(doc)
  for w in features:
    curr_features[w]=w in words_set
  return curr_features

train_data=[]
for i in range(len(train_docs)):
  train_data.append((get_feature_dict(train_docs[i]),y_train[i]))

test_data=[]
for i in range(len(test_docs)):
  test_data.append(get_feature_dict(test_docs[i]))

```

```python
print(len(train_data[0][0]))
print(len(test_data[0]))

print(type(test_data[0]))
```

```python
##classification using nltk
from nltk import NaiveBayesClassifier
clf=NaiveBayesClassifier.train(train_data)
```

```python
y_test=[]
for i in range(len(test_data)):
  y_test.append(clf.classify(test_data[i]))
```
